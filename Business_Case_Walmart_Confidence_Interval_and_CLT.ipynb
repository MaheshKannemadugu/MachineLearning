```python
from scipy.stats import norm

def confidence_interval(sample, confidence=0.95):
    mean = np.mean(sample)
    std = np.std(sample, ddof=1)
    n = len(sample)
    z = norm.ppf(1 - (1-confidence)/2)
    margin = z * (std/np.sqrt(n))
    return mean - margin, mean + margin, mean
```

```python
male = df[df['Gender']=='M']['Purchase']
female = df[df['Gender']=='F']['Purchase']

confidence_interval(male, 0.95)
confidence_interval(female, 0.95)
```

```python
#Central Limit Theorem Demonstration
means = []
for i in range(5000):
    sample = female.sample(100, replace=True)
    means.append(sample.mean())

sns.histplot(means, kde=True)
```

```python
male = df[df["Gender"]=="M"]["Purchase"]
female = df[df["Gender"]=="F"]["Purchase"]

print("Male 95% CI:", confidence_interval(male, 0.95))
print("Female 95% CI:", confidence_interval(female, 0.95))
```

```python
print("Male 90% CI:", confidence_interval(male, 0.90))
print("Female 90% CI:", confidence_interval(female, 0.90))

print("Male 99% CI:", confidence_interval(male, 0.99))
print("Female 99% CI:", confidence_interval(female, 0.99))
```

```python
means_f = []
for i in range(5000):
    sample = female.sample(100, replace=True)
    means_f.append(sample.mean())

sns.histplot(means_f, kde=True)
plt.title("Sampling Mean Distribution - Female")
plt.show()
```

```python
# Married vs Unmarried
married = df[df["Marital_Status"]==1]["Purchase"]
unmarried = df[df["Marital_Status"]==0]["Purchase"]

df.groupby("Marital_Status")["Purchase"].mean()
```
