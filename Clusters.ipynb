```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans, DBSCAN
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster

# Generate synthetic data
X, _ = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# K-Means Clustering
def kmeans_clustering(X):
    kmeans = KMeans(n_clusters=4)
    kmeans.fit(X)
    centers = kmeans.cluster_centers_
    labels = kmeans.labels_

    plt.figure(figsize=(12, 4))
    plt.subplot(131)
    plt.scatter(X[:, 0], X[:, 1], c=labels, s=50, cmap='viridis')
    plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker='X') # Cluster centers
    plt.title('K-Means Clustering')

# DBSCAN Clustering
def dbscan_clustering(X):
    dbscan = DBSCAN(eps=0.6, min_samples=10)
    labels = dbscan.fit_predict(X)

    plt.subplot(132)
    plt.scatter(X[:, 0], X[:, 1], c=labels, s=50, cmap='viridis')
    plt.title('DBSCAN Clustering')

# Hierarchical Clustering
def hierarchical_clustering(X):
    linked = linkage(X, method='ward')
    labels = fcluster(linked, t=4, criterion='maxclust')

    plt.subplot(133)
    plt.scatter(X[:, 0], X[:, 1], c=labels, s=50, cmap='viridis')
    plt.title('Hierarchical Clustering')

    plt.figure(figsize=(10, 7))
    dendrogram(linked, truncate_mode='lastp', p=12, show_leaf_counts=False, color_threshold=0.7*max(linked[:,2]))
    plt.title('Dendrogram')
    plt.xlabel('Cluster Size')
    plt.ylabel('Distance')
    plt.show()

# Execute clustering functions
kmeans_clustering(X)
dbscan_clustering(X)
hierarchical_clustering(X)

plt.show()
```
